```python
import pandas as pd
import matplotlib.pyplot as plot
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv(r"/Users/sreejagouni/Desktop/cleaned_data.csv")
print(data)
```

```python
sns.scatterplot(x='Age', y='Grip Strength', data=data)
plot.title('Age Vs Grip Strength')
plot.xlabel('Age')
plot.ylabel('Grip Strength')
plot.savefig(r"/Users/sreejagouni/Desktop/agevsgrip.png")
plot.show()
```

```python
sns.barplot(x='Height', y='Grip Strength', data=data)
plot.title('Height Vs Grip Strength')
plot.xlabel('Height')
plot.ylabel('Grip Strength')
plot.savefig(r"/Users/sreejagouni/Desktop/heightvsgrip.png")
plot.show()
```

```python
matrix = data[['Age','Height','Weight','Grip Strength']].corr()
```

```python
sns.heatmap(matrix, annot=True, cmap= 'coolwarm')
plot.title('Heat Map')
plot.savefig(r"/Users/sreejagouni/Desktop/heatmap.png")
plot.show()
```
