```python
import pandas as pd
import matplotlib.pyplot as plot
import numpy as np
import seaborn as sns
```

```python
data = pd.read_csv(r"/Users/sreejagouni/Downloads/StudentsPerformance_cleaned_data.csv")
```

```python
print(data)
```

```python
lunchCount = data['lunch'].value_counts()
print(lunchCount)
```

```python
plot.figure(figsize=(6,6))
plot.pie(lunchCount, labels=lunchCount.index)
plot.title("Types of Lunch Distribution")
plot.savefig(r"/Users/sreejagouni/Desktop/lunchdistribution.png")
plot.show()
```

```python
plot.scatter(data['reading score'], data['writing score'])
plot.xlabel('Reading Score')
plot.ylabel('Writing Score')
plot.title("Reading VS Writing Scores")
plot.savefig(r"/Users/sreejagouni/Desktop/readingvswriting.png")
plot.show()
```

```python
sns.barplot(x='gender', y='math score', data=data)
plot.xlabel('Gender')
plot.ylabel('Math Score')
plot.title("Math Score By Gender")
plot.savefig(r"/Users/sreejagouni/Desktop/mathscorebygender.png")
plot.show()
```

```python
sns.boxplot(x='gender', y='reading score', data=data)
plot.xlabel('Gender')
plot.ylabel('Reading Score')
plot.title("Reading Score By Gender")
plot.savefig(r"/Users/sreejagouni/Desktop/readingscorebygender.png")
plot.show()
```

```python
plot.hist(data['writing score'], bins=30, color='pink',edgecolor='blue')
plot.xlabel('Writing Score')
plot.ylabel('Numbers')
plot.title("Writing Score Numbers")
plot.savefig(r"/Users/sreejagouni/Desktop/writingscorenumbers.png")
plot.show()
```
